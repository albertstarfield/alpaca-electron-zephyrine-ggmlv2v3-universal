```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, AutoConfig
import accelerate


# Oh yeah btw for now the GPU is cuda since the full implementation as always is Nvidia
#https://stackoverflow.com/questions/77547377/notimplementederror-cannot-copy-out-of-meta-tensor-no-data


#Cloud you piece of shit
modelNamePath = "Open-Orca/Mistral-7B-OpenOrca"

#from huggingface_hub import snapshot_download
#snapshot_download(repo_id=modelNamePath, local_dir="Mistral-7B-OpenOrca",
#                  local_dir_use_symlinks=False, revision="main")

#---------------------------
# Use Local / FUCK CENTRALIZED CLOUD
# https://stackoverflow.com/questions/64001128/load-a-pre-trained-model-from-disk-with-huggingface-transformers
modelNamePath = "/Users/user/adelaide-zephyrine-charlotte-assistant/usr/engine_component/generalReintegrationPyWorkstationEnv/LLMReintegrationExperiment/huggingface_cache/models--Open-Orca--Mistral-7B-OpenOrca/snapshots/4a37328cef00f524d3791b1c0cc559a3cc6af14d"

#BitsandBytes is CUDA Exclusive Nvidia Only, Excluded from the main code
"""bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
    #llm_int8_enable_fp32_cpu_offload=True
)"""

## Load the Model into Machine?
## Can we reduce the memory usage, since there will be an instance of LLM running on the

#https://huggingface.co/docs/transformers/en/quantization
from transformers import QuantoConfig
quantization_config = QuantoConfig(weights="int8")


#CPU Memory Management
import psutil
def get_available_memory_mb():
  """
  This function retrieves the available allocatable memory on the system
  and returns it in Megabytes (MB).
  """
  # Get total available memory in bytes
  total_memory = psutil.virtual_memory().available

  # Divide by 1024**2 to convert bytes to Megabytes (MB)
  # and round down to the nearest whole number using //
  available_memory_mb = total_memory // (1024 * 1024)

  # Divide available memory by 2
  half_available_memory_mb = available_memory_mb // 2
  print(f'Capping Memory Usage {half_available_memory_mb} MB')
  return half_available_memory_mb

memory_mb = get_available_memory_mb()

# To prevent buffer issues (with the training_swapcache) we need to initialize with empty 
#https://github.com/huggingface/transformers/issues/28972#issuecomment-1939089314
#Nope still empty tensors

#Lets try this
#https://zhuanlan.zhihu.com/p/646554710 #Some random Forum
#https://zhuanlan.zhihu.com/p/656307174 #Basic Environment QnA

from accelerate import init_empty_weights


config = AutoConfig.from_pretrained(modelNamePath)

with init_empty_weights():
    model = AutoModelForCausalLM.from_pretrained(
        modelNamePath,
        #quantization_config=bnb_config, #Special only for Nvidia, which our target is universal system so it's not doeable on Zephy
        trust_remote_code=True,
        #torch_dtype=torch.float16, #Auto will cause it to enforce Nvidia only support GPU training
        torch_dtype=torch.float16,
        use_cache=False,
        device_map="auto", #Broken Implementation causes a empty tensors error, # device map{'': 'disk'}
        max_memory={"cpu": f"{memory_mb}MB"}, #https://github.com/huggingface/accelerate/issues/483 also we're going to implement GPU limit later
        #quantization_config=quantization_config,
        offload_state_dict=True,
        #attn_implementation="flash_attention_2", #Failure on a specific platform compilation, will try to avoid it
        offload_folder="./training_swapcache" #disk swapping for training (Since we're on a situation where 80% is allocated by something... or inference or smth)
    ) #https://github.com/huggingface/accelerate/issues/2129 Forces everything to CPU




#model.config.use_cache = False
#model.is_parallelizable = True
#model.model_parallel = False


#Load The Tokenizer either for read (Inferencing)/ write (Training)

tokenizer = AutoTokenizer.from_pretrained(modelNamePath, trust_remote_code=True) #Grabbed from the repo not from the main memory of the model
tokenizer.pad_token = tokenizer.eos_token #Padding


from accelerate import infer_auto_device_map, init_empty_weights
infer_auto_device_map(model)
```

```python
def correct_answer(ans):
  return (ans.split("->:")[1]).strip()

answers = []
for ix,seq in enumerate(sequences):
    # print(ix,seq[0]['generated_text'])
    answers.append(correct_answer(seq[0]['generated_text']))

answers
```
